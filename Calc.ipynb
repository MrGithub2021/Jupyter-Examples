```python
x = input("What is your name?")
print("Hi " + x + "!")
# My Calcuator!
from ipywidgets import widgets
lbl1 = widgets.Label("Type the first number:")
display(lbl1)
text1 = widgets.Text()
display(text1)
lbl2 = widgets.Label("Type the second number:")
display(lbl2)
text2= widgets.Text()
display(text2)
btn0 = widgets.Button(description = "Add them!")
display(btn0)
btn1 = widgets.Button(description = "Subtract them!")
display(btn1)
btn2 = widgets.Button(description = "Multiply them!")
display(btn2)
btn3 = widgets.Button(description = "Divide them!")
display(btn3)
btn4 = widgets.Button(description = "Get the remainder!!")
display(btn4)
lbl3 = widgets.Label()
display(lbl3)
def add(b):
    x=int(text1.value)
    y=int(text2.value)
    lbl3.value = "The result is: " + str(x+y)
btn0.on_click(add) 
def sub(b):
    x=int(text1.value)
    y=int(text2.value)
    lbl3.value = "The result is: " + str(x-y)
btn1.on_click(sub)
def mul(b):
    x=int(text1.value)
    y=int(text2.value)
    lbl3.value = "The result is: " + str(x * y)
btn2.on_click(mul)
def div(b):
    x=int(text1.value)
    y=int(text2.value)
    lbl3.value = "The result is: " + str(x/y)
btn3.on_click(div)
def rem(b):
    x=int(text1.value)
    y=int(text2.value)
    lbl3.value = "The result is: " + str(x%y)
btn4.on_click(rem)
```
